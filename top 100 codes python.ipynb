```python
#LCM of two numbers:
#pyhton:
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

def gcd(a, b):
    # Euclidean algorithm
    while b != 0:
        a, b = b, a % b
    return a

def lcm(a, b):
    return (a * b) // gcd(a, b)

print(f"LCM of {num1} and {num2} is {lcm(num1, num2)}")
```

```python
#Can a number be expressed as a sum of two prime numbers :
#python:
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter any number: "))

found = False
for i in range(2, num):
    if is_prime(i) and is_prime(num - i):
        print(f"{num} can be expressed as {i} + {num - i}")
        found = True
        break

if not found:
    print(f"{num} cannot be expressed as the sum of two prime numbers")
```

```python
#Counting number of days in a given month of a year
#python:
def leapyear_validation(year):
    return (year%4 == 0 and year%100 != 0) or (year%400 == 0)
        
month = input("Enter month: ").lower()
year = int(input("Enter year: "))
no_of_days = {"january":31,"february":28,"march":31,"april":30,"may":31,"june":30,"july":31,"agust":31,"september":30,"october":31,"november":30,"december":31}
if month in no_of_days:
    if month == "february" and leapyear_validation(year):
        print(f"no.of days = 29")
    else:
        print(f"No .of days = {no_of_days[month]}")
else:
    print("invalid month entered")
```

```python
#Finding Number of times x digit occurs in a given input
#python:
num = int(input("Enter any number: "))
x = int(input("Enter any digit: "))
count = 0
for digit in str(num):
    if int(digit) == x :
        count += 1
print(count)
```

```python
#Finding number of integers which has exactly x divisors
#python:
x = int(input("Enter no.of divisors: "))
n = int(input())
count = 0
for i in range(1,n+1):
    factors = 0
    for j in range(1,i+1):
        if i%j == 0 :
            factors += 1
        else:
            pass
    if factors == x:
        count += 1
print(count)
```
