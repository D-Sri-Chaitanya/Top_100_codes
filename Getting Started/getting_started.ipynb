```python
#prime number or not
#python:
num = int(input("Enter any number: "))
if num < 2 :
    print(f"{num} is not prime")
else:
    is_prime = True
    for i in range(2,int(num**0.5)+1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"given number : {num} is prime")
    else:
        print(f"given number : {num} is not prime")
```

```python
#prime number in a range
#python:
start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))
for p in range(start,end+1):
    if p < 2 :
        continue
    is_prime = True
    for i in range(2,int(p**0.5)+1):
        if p % i == 0:
            is_prime = False
            break
    if is_prime:
        print(p,end = " ")
```

```python
#armstrong number
#python:
num = int(input("Enter any number"))
temp = num
no_of_digits = len(str(num))
sum = 0
for i in range(no_of_digits):
    sum += (num % 10) ** no_of_digits
    num = num // 10
if sum == temp:
    print(f"given number is armstrong number")
else:
    print(f"given number is not armstrong number")
```

```python
# find nth term of fibonacci series:
# python:
n = int(input("Enter any position: "))
if n < 0 :
    print("invalid")
elif n == 0:
    print(f"{n}th term in fibonacci series is {0}")
elif n == 1:
    print(f"{n}th term in fibonacci series is {1}")
else:
    a,b = 0,1
    for i in range(2,n+1):
       a,b = b,a+b
    print(f"{n} th term in fibonacci series is {b} ")
```

```python
#strong number:
#python:
num = int(input("Enter any number: "))
temp = num
sum = 0
while num > 0:
    digit = num % 10
    fact = 1
    for j in range(1,digit+1):
        fact *= j
    sum = sum + fact
    num //= 10
if sum == temp:
    print(f"the given number : {temp} is strong number")
else:
    print(f"the given number : {temp} is not strong number")   
```

```python
#perfect number
#python:
num = int(input("Enter any number: "))
sum = 0
for i in range(1,num):
    if num % i == 0:
        sum += i
    else:
        continue
if sum == num:
    print(f"the given number : {num} is perfect")
else:
    print(f"the given number : {num} is not perfect")
```

```python
#automorphic number:
#python:
num = int(input("Enter any number: "))
square = num ** 2
lastdigits = str(square)[-len(str(num)):]
if lastdigits == str(num):
    print("given number is automorphic")
else :
    print("given number is not automorphic")
```

```python
#perfect square:
#python:
num = int(input("Enter any number: "))
for i in range(1,num):
    if num % i == 0:
        square = i**2
    else:
        continue
if square == num:
    print("perfect square")
else :
    print("not a perfect square")
```

```python
#harshad number
#python:
num = int(input("Enter any number: "))
temp = num
sum = 0
while num != 0:
    sum += num % 10
    num //= 10
if temp % sum == 0:
    print("is a harshad number")
else:
    print("not a harshad number")
```

```python
#abundant number:
#python:
num = int(input("Enter any number: "))
sum = 0
for i in range(1,num):
    if num % i == 0:
        sum += i
    else:
        continue
if sum > num:
    print("abundant")
else:
    print("not abundant")
```

```python
#friendly pair
#python:
# Friendly Pair in Python

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

def sum_of_divisors(n):
    total = 0
    for i in range(1, n+1):
        if n % i == 0:
            total += i
    return total

sum1 = sum_of_divisors(num1)
sum2 = sum_of_divisors(num2)

if sum1 // num1 == sum2 // num2:
    print("Friendly pair")
else:
    print("Not a friendly pair")
```
