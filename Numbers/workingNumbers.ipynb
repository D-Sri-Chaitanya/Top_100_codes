```python
#LCM of two numbers:
#pyhton:
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

def gcd(a, b):
    # Euclidean algorithm
    while b != 0:
        a, b = b, a % b
    return a

def lcm(a, b):
    return (a * b) // gcd(a, b)

print(f"LCM of {num1} and {num2} is {lcm(num1, num2)}")
```

```python
#binary to octal:
#python:
def binary_validation(binary_num):
    for digit in binary_num:
        if digit not in "01":
            return False
    return True
num = input("enter any number: ")
# if binary_validation(num):
#     octal_num = oct(int(num, 2))[2:]
#     print("Octal equivalent:", octal_num)
# else:
#     print("Invalid binary number!")
def binary_to_octal(binary_num):
    while len(binary_num) % 3 != 0:
        binary_num = "0" + binary_num
    octal = ""
    for i in range(0, len(binary_num), 3):
        group = binary_num[i:i+3]
        octal += str(int(group, 2))  
    return octal
print(binary_to_octal(num))  
```

```python
#Quadrants in which a given coordinate lies
#python:
def quadrants(x,y):
    if x > 0 and y > 0:
        print("the given coordinates are in quadrant 1")
    elif x < 0 and y > 0:
        print("the given coordinates are in quadrant 2")
    elif x < 0 and y < 0:
        print("the given coordinates are in quadrant 3")
    else:
        print("the given coordinates are in quadrant 4")
x,y = map(int,input("enter any two numbers: ").split(" "))
quadrants(x,y)
```

```python
#Can a number be expressed as a sum of two prime numbers :
#python:
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter any number: "))

found = False
for i in range(2, num):
    if is_prime(i) and is_prime(num - i):
        print(f"{num} can be expressed as {i} + {num - i}")
        found = True
        break

if not found:
    print(f"{num} cannot be expressed as the sum of two prime numbers")
```
