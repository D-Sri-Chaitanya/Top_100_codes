```python
#power of a number using recursion:
def power_of_number(n, p):
    if p == 0:
        return 1
    return n * power_of_number(n, p - 1)

n = int(input("Enter the number: "))
p = int(input("Enter the power: "))
power = power_of_number(n, p)
print(power)
```

```python
#prime number using recursion:
def prime_number(n,divisor = 2):
    if n <= 1:
        return "not prime"
    if divisor > int(n**0.5):
        return "prime"
    if n % divisor == 0 :
        return "not prime"
    return prime_number(n,divisor+1)
n = int(input("enter n value : "))
print(prime_number(n))            
```

```python
#smallest element in an array using recursion:
def find_min(arr,left,right):
    if left == right:
        return arr[left]
    mid = (left+right)//2
    min_left = find_min(arr,left,mid)
    min_right = find_min(arr,mid+1,right)
    if min_left < min_right:
        return min_left
    else:
        return min_right
arr = list(map(int,input("Enter array values: ").split(",")))
print("smallest element:",find_min(arr,0,len(arr)-1))
```

```python
#reversing a number using recursion:
def reverse_number(num,rev = 0):
    if num == 0:
        return rev
    last_digit = num % 10
    rev = (rev * 10) + last_digit
    return reverse_number(num//10,rev)
num = int(input("Enter any number: "))
print(reverse_number(num))
```

```python
# HCF of two numbers using recursion:
def HCF(num1,num2):
    if num2 == 0:
        return num1
    else:
        return HCF(num2,num1%num2)
num1 = int(input())
num2 = int(input())
print(HCF(num1,num2))
```

```python
# LCM of two numbers using recursion:
def HCF(num1,num2):
    if num2 == 0:
        return num1
    else:
        return HCF(num2,num1%num2)
def LCM(num1,num2):
    return (num1*num2)//HCF(num1,num2)
num1 = int(input())
num2 = int(input())
print(LCM(num1,num2))
```

```python
#printing all the permutations of a string using recursion:
def permute(s,answer = ""):
    if len(s) == 0:
        print(answer)
        return
    for i in range(len(s)):
        ch = s[i]
        left_str = s[:i]
        right_str = s[i+1:]
        remaining = left_str + right_str
        permute(remaining,answer + ch)
s = input()
permute(s)
```

```python
# Given an integer N the task is to print the F(N)th term:
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)
N = int(input("Enter N: "))
print("F(", N, ") =", fibonacci(N))
```

```python
#Given a list arr of N integers, print sums of all subsets in it using recursion:
def subset_sums(arr):
    result = []
    def solve(index, current_sum):
        if index == len(arr):
            result.append(current_sum)
            return
        solve(index + 1, current_sum + arr[index])
        solve(index + 1, current_sum)
    solve(0, 0)
    return result
arr = list(map(int,input().split(",")))
print(subset_sums(arr))
```

```python
#Last non-zero digit in factorial using recursion:
def last_non_zero_digit(n, result=1):
    if n == 0:
        return result % 10
    result = result * n
    while result % 10 == 0:
        result = result // 10
    return last_non_zero_digit(n-1, result)
n = int(input())
print(last_non_zero_digit(n))
```

```python
#Given a positive integer N, return the Nth row of pascal’s triangle using recursion:
#brute force:
# def pascal_value(n, k):
#     if k == 0 or k == n:
#         return 1
    
#     return pascal_value(n-1, k-1) + pascal_value(n-1, k)

# def pascal_row(n):
#     row = []
#     for k in range(n + 1):
#         row.append(pascal_value(n, k))
#     return row

#optimized
def pascal_row(n):
    if n == 0:
        return [1]
    prev = pascal_row(n - 1)
    row = [1]
    for i in range(1, n):
        row.append(prev[i-1] + prev[i])
    row.append(1)
    return row

n = int(input("Enter row number: "))
print(pascal_row(n))
```

```python
#Given an integer N representing the number of pairs of parentheses, the task is to generate all combinations of well-formed(balanced) parentheses using recursion:
def generate_parentheses(n):
    result = []
    def backtrack(curr, open_count, close_count):
        if len(curr) == 2 * n:
            result.append(curr)
            return
        if open_count < n:
            backtrack(curr + "(", open_count + 1, close_count)
        if close_count < open_count:
            backtrack(curr + ")", open_count, close_count + 1)
    backtrack("", 0, 0)
    return result
n = int(input("Enter number of pairs: "))
print(generate_parentheses(n))
```

```python
# factorial of a number using recursion:
def factorial(n):
    if n == 0:
        return 1
    if n == 1:
        return 1
    return n*factorial(n-1)
n = int(input())
print(factorial(n))
```

```python
# N-bit binary numbers having more  or equal 1’s than 0’s using recursion:
def printRec(number, extraOnes, remainingPlaces):
    if 0 == remainingPlaces:
        print(number, end=" ")
        return

    printRec(number + "1", extraOnes + 1, remainingPlaces - 1)

    if 0 < extraOnes:
        printRec(number + "0", extraOnes - 1, remainingPlaces - 1)

def printNums(n):
    str = ""
    printRec(str, 0, n)

n = int(input())
printNums(n)
```

```python
# Given a string s, remove all its adjacent duplicate characters recursively :
def remove_duplicates(s):
    if len(s) <= 1:
        return s

    if s[0] == s[1]:

        while len(s) > 1 and s[0] == s[1]:
            s = s[1:]

        return remove_duplicates(s[1:])

    else:
        return s[0] + remove_duplicates(s[1:])


s = input()
print(remove_duplicates(s))
```
