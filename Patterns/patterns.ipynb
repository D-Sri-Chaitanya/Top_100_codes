```python
#patterns:
# square star pattern:
n = int(input())
for i in range(n):
    for j in range(n):
        print("*", end = " ")
    print()
```

```python
# Hollow square pattern:
n = int(input("Enter size: "))
for i in range(1, n+1):          
    for j in range(1, n+1):     
        
        if i == 1 or i == n or j == 1 or j == n:
            print("*", end=" ")
        else:
            print(" ", end=" ")
            
    print()
```
